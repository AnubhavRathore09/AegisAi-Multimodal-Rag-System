# AegisAI Multimodal RAG Demo

This notebook demonstrates:
- Adaptive routing (direct / rag / memory)
- API interaction
- Evaluation system


```python
import requests

BASE_URL = "http://127.0.0.1:8000"
```

```python
url = f"{BASE_URL}/api/chat"

payload = {"message": "What is artificial intelligence?"}

response = requests.post(url, json=payload)
response.json()
```

```python
payload = {"message": "Explain machine learning from document"}

response = requests.post(url, json=payload)
response.json()
```

```python
payload = {"message": "What did I ask before?"}

response = requests.post(url, json=payload)
response.json()
```

```python
url = f"{BASE_URL}/api/evaluate"

payload = {
  "query": "What is AI?",
  "response": "AI is artificial intelligence"
}

response = requests.post(url, json=payload)
response.json()
```
